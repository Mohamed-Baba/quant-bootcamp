````python
# ============================================================================
# EXERCICE 6 : Alg√®bre lin√©aire (Portfolio theory)
# ============================================================================
print("\n--- Exercice 6 : Linear Algebra pour Finance ---")

# Rendements de 3 actions
returns = np.array([0.05, 0.08, 0.03])

# Poids du portfolio
weights = np.array([0.3, 0.5, 0.2])

# TODO: Calcule le rendement du portfolio
# Formule: sum(w_i * r_i) = w^T * r
portfolio_return = np.dot(returns , weights)  # Astuce: np.dot()    # C'est le produit matriciel 

print(f"\nRendement portfolio: {portfolio_return:.2%}")

# Matrice de covariance (simplifi√©)
cov_matrix = np.array([
    [0.04, 0.01, 0.02],
    [0.01, 0.09, 0.015],
    [0.02, 0.015, 0.03]
])

# TODO: Calcule la variance du portfolio
# Formule: w^T * Cov * w
portfolio_variance = weights @ cov_matrix @ weights  # Astuce: w.T @ Cov @ w

# TODO: Calcule la volatilit√© (√©cart-type)
portfolio_volatility = np.sqrt(portfolio_variance)

print(f"Volatilit√© portfolio: {portfolio_volatility:.2%}")


print("\n" + "="*70)
print("JOUR 3 : MANIPULATION AVANC√âE & BOOLEAN INDEXING")
print("="*70)
````

````python
# ============================================================================
# EXERCICE 10 : Rolling windows (crucial pour backtesting)
# ============================================================================
print("\n--- Exercice 10 : Rolling Statistics ---")

prices_series = 100 * np.exp(np.cumsum(np.random.randn(100) * 0.02))

# TODO: Calcule une moyenne mobile sur 10 jours
# Astuce: utilise np.convolve ou une boucle intelligente
window = 10
moving_average = np.convolve(prices_series, np.ones(window)/window, mode='valid')

# TODO: Calcule la volatilit√© roulante (rolling std) sur 20 jours
# C'est plus complexe, mais essaie !
window = 20
rolling_std = np.array([np.std(prices_series[i:i+window]) 
                        for i in range(len(prices_series) - window + 1)])

print(f"\nPrix initial: {prices_series[0]:.2f}")
print(f"Prix final: {prices_series[-1]:.2f}")
print(f"MA(10) final: {moving_average[-1]:.2f}")


print("\n" + "="*70)
print("JOUR 5 : SIMULATION MONTE CARLO")
print("="*70)

# ============================================================================
# EXERCICE 11 : Simulation de trajectoires (GBM)
# ============================================================================
print("\n--- Exercice 11 : Geometric Brownian Motion ---")

# Param√®tres
S0 = 100        # Prix initial
mu = 0.10       # Drift (rendement annuel)
sigma = 0.20    # Volatilit√© annuelle
T = 1.0         # Horizon (1 an)
n_steps = 252   # Nombre de pas (jours de trading)
n_sims = 1000   # Nombre de simulations

dt = T / n_steps
np.random.seed(42)

# TODO: Simule n_sims trajectoires de prix
# Formule: S_{t+dt} = S_t * exp((mu - 0.5*sigma^2)*dt + sigma*sqrt(dt)*Z)
# O√π Z ~ N(0,1)

# M√©thode 1 : Boucle (pour comprendre)
paths = np.zeros((n_steps + 1, n_sims))
paths[0] = S0

for t in range(1, n_steps + 1):
    Z = np.random.randn(n_sims)
    paths[t] = paths[t-1] * np.exp((mu - 0.5*sigma**2)*dt + sigma*np.sqrt(dt)*Z)

# TODO: Calcule le prix moyen √† maturit√©
mean_terminal_price = None

# TODO: Calcule l'intervalle de confiance √† 95%
conf_interval = None  # Astuce: percentiles 2.5% et 97.5%

print(f"\n*** R√âSULTATS SIMULATION ***")
print(f"Prix moyen √† T={T}: {mean_terminal_price:.2f}")
print(f"Intervalle 95%: [{conf_interval[0]:.2f}, {conf_interval[1]:.2f}]")
print(f"Prix th√©orique: {S0 * np.exp(mu * T):.2f}")

# Visualisation
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(paths[:, :20], alpha=0.5, linewidth=0.8)  # 20 premi√®res trajectoires
plt.axhline(S0, color='red', linestyle='--', label='Prix initial')
plt.xlabel('Jours')
plt.ylabel('Prix')
plt.title('Trajectoires GBM')
plt.legend()
plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
plt.hist(paths[-1], bins=50, alpha=0.7, edgecolor='black')
plt.axvline(mean_terminal_price, color='red', linestyle='--', label='Moyenne')
plt.xlabel('Prix terminal')
plt.ylabel('Fr√©quence')
plt.title('Distribution des prix √† maturit√©')
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('gbm_simulation.png', dpi=150)
print("\nGraphique sauvegard√©: gbm_simulation.png")
plt.show()
````

````python

# ============================================================================
# EXERCICE 14 : Backtesting d'une strat√©gie
# ============================================================================
print("\n--- Exercice 14 : Simple Moving Average Strategy ---")

# G√©n√®re des prix fictifs
np.random.seed(42)
n_days = 500
prices = 100 * np.exp(np.cumsum(np.random.randn(n_days) * 0.015))

# Param√®tres strat√©gie
short_window = 20
long_window = 50

# TODO: Calcule les moyennes mobiles
short_ma = np.convolve(prices, np.ones(short_window)/short_window, mode='valid')
long_ma = np.convolve(prices, np.ones(long_window)/long_window, mode='valid')

# Ajuste les longueurs
prices_aligned = prices[long_window-1:]
short_ma_aligned = short_ma[long_window-short_window:]

# TODO: G√©n√®re les signaux de trading
# Signal = 1 si short_ma > long_ma (achat), -1 sinon (vente)
signals = np.where(short_ma_aligned > long_ma, 1, -1)

# TODO: Calcule les rendements de la strat√©gie
daily_returns = np.diff(prices_aligned) / prices_aligned[:-1]
strategy_returns = signals[:-1] * daily_returns

# TODO: Calcule les performances
cumulative_returns = np.cumprod(1 + strategy_returns) - 1
total_return = cumulative_returns[-1]
sharpe_ratio = np.mean(strategy_returns) / np.std(strategy_returns) * np.sqrt(252)

# Drawdown
cumulative = np.cumprod(1 + strategy_returns)
running_max = np.maximum.accumulate(cumulative)
drawdown = (cumulative - running_max) / running_max
max_drawdown = np.min(drawdown)

print(f"\n*** R√âSULTATS BACKTEST ***")
print(f"Rendement total: {total_return:.2%}")
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Max Drawdown: {max_drawdown:.2%}")
print(f"Nombre de trades: {np.sum(np.abs(np.diff(signals)))}")

# Visualisation
fig, axes = plt.subplots(3, 1, figsize=(12, 10))

# Prix et moyennes mobiles
axes[0].plot(prices_aligned, label='Prix', alpha=0.7)
axes[0].plot(short_ma_aligned, label=f'MA{short_window}', linewidth=2)
axes[0].plot(long_ma, label=f'MA{long_window}', linewidth=2)
axes[0].set_ylabel('Prix')
axes[0].set_title('Prix et Moyennes Mobiles')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Signaux
axes[1].fill_between(range(len(signals)), 0, signals, alpha=0.3, 
                     label='Signal (1=Long, -1=Short)')
axes[1].set_ylabel('Position')
axes[1].set_title('Signaux de Trading')
axes[1].legend()
axes[1].grid(alpha=0.3)

# Equity curve
axes[2].plot(cumulative_returns * 100, label='Strat√©gie', linewidth=2)
axes[2].axhline(0, color='black', linestyle='--', alpha=0.5)
axes[2].set_xlabel('Jours')
axes[2].set_ylabel('Rendement cumul√© (%)')
axes[2].set_title('Performance de la Strat√©gie')
axes[2].legend()
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('backtest_results.png', dpi=150)
print("\nGraphique sauvegard√©: backtest_results.png")
plt.show()


print("\n" + "="*70)
print("F√âLICITATIONS ! TU AS TERMIN√â LE BOOTCAMP NUMPY !")
print("="*70)
print("""
PROCHAINES √âTAPES :
1. Refais tous les exercices de m√©moire (sans regarder)
2. Impl√©mente des variations (change les param√®tres, les strat√©gies)
3. Passe au projet Option Pricer que je t'ai donn√© avant
4. Commence √† lire "Python for Finance" de Yves Hilpisch

TU ES MAINTENANT PR√äT √Ä CONQU√âRIR LE MONDE QUANT !

N'oublie pas :
- Code TOUS LES JOURS
- Push sur GitHub CHAQUE SEMAINE
- Impl√©mente des papers dans 2-3 semaines
- Dans 3 mois, tu seras IMBATTABLE

MAINTENANT, AU TRAVAIL ! üí™üî•
""")
````
